```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
import seaborn as sns

# Configuration esthétique
plt.style.use('seaborn-v0_8-darkgrid')

# =============================================================================
# 1. FONCTION DE CHARGEMENT VIA API (PLUS STABLE)
# =============================================================================

def charger_opendata_paris(dataset_id, rows=1000):
    """Charge un dataset via l'API Open Data Paris"""
    base_url = f"https://opendata.paris.fr/api/records/1.0/search/?dataset={dataset_id}&rows={rows}"
    
    try:
        response = requests.get(base_url)
        if response.status_code == 200:
            data = response.json()
            # On extrait les champs 'fields' de chaque enregistrement
            records = [r['fields'] for r in data['records']]
            df = pd.DataFrame(records)
            print(f"✅ Dataset '{dataset_id}' chargé ({len(df)} lignes).")
            return df
        else:
            print(f"❌ Erreur {response.status_code} pour le dataset {dataset_id}")
            return None
    except Exception as e:
        print(f"❌ Erreur de connexion : {e}")
        return None

# =============================================================================
# 2. CHARGEMENT DES DEUX JEUX DE DONNÉES
# =============================================================================

print("🚀 Lancement du chargement des données depuis Open Data Paris...")

df_sorties = charger_opendata_paris("que-faire-a-paris-")
df_chantiers = charger_opendata_paris("chantiers-perturbants")


display(df_sorties.head())

display(df_chantiers.head())

# =============================================================================
# 3. EXPLORATION SÉCURISÉE DES SORTIES (ÉVÉNEMENTS)
# =============================================================================

if df_sorties is not None:
    print("\n--- 🎭 ANALYSE DES SORTIES À PARIS ---")
    
    # Liste des colonnes qu'on aimerait avoir
    cols_interet = ['title', 'category_label', 'address_zipcode', 'tags', 'date_start']
    # On ne garde que celles qui existent vraiment pour éviter le KeyError
    cols_valides = [c for c in cols_interet if c in df_sorties.columns]
    
    print(f"Colonnes utilisées : {cols_valides}")
    display(df_sorties[cols_valides].head())

    # Petit graphique : Top codes postaux pour les événements
    if 'address_zipcode' in df_sorties.columns:
        plt.figure(figsize=(10, 5))
        df_sorties['address_zipcode'].value_counts().head(10).plot(kind='bar', color='skyblue')
        plt.title("Top 10 des arrondissements par nombre d'événements")
        plt.ylabel("Nombre d'activités")
        plt.show()

# =============================================================================
# 4. EXPLORATION SÉCURISÉE DES CHANTIERS
# =============================================================================

if df_chantiers is not None:
    print("\n--- 🏗️ ANALYSE DES CHANTIERS PERTURBANTS ---")
    
    # Colonnes probables pour les chantiers
    cols_chantier_interet = ['objet', 'niveau_perturbation', 'cp_arrondissement', 'impact_circulation']
    cols_chantier_valides = [c for c in cols_chantier_interet if c in df_chantiers.columns]
    
    print(f"Colonnes utilisées : {cols_chantier_valides}")
    display(df_chantiers[cols_chantier_valides].head())

    # Analyse de l'impact sur la circulation
    if 'impact_circulation' in df_chantiers.columns:
        print("\n📊 Impact sur la circulation :")
        print(df_chantiers['impact_circulation'].value_counts())
        
        plt.figure(figsize=(10, 5))
        sns.countplot(data=df_chantiers, y='impact_circulation', palette="OrRd_r")
        plt.title("Répartition de l'impact des chantiers sur la circulation")
        plt.show()
```

# Événements et chantiers à Paris — via l'API Open Data v2.1

Migration depuis l'ancienne API `records/1.0/search/` (dépréciée) vers la nouvelle **API Explore v2.1** :
`https://opendata.paris.fr/api/explore/v2.1/catalog/datasets/{dataset_id}/records`

Différences principales :
- Nouveau endpoint : `/api/explore/v2.1/catalog/datasets/{id}/records`
- Paramètre `limit` (max 100 par requête) au lieu de `rows`
- Pagination via `offset` pour récupérer plus de 100 lignes
- Les champs sont directement à la racine de chaque record (plus besoin d'extraire `fields`)

```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
import seaborn as sns

# Configuration esthétique
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette('pastel')
```

```python
# =============================================================================
# 1. FONCTION DE CHARGEMENT VIA L'API EXPLORE v2.1 (avec pagination)
# =============================================================================

def charger_opendata_paris(dataset_id, total_max=1000, page_size=100, timeout=15):
    """
    Charge un dataset via l'API Explore v2.1 d'Open Data Paris.

    Args:
        dataset_id (str): identifiant du dataset (ex: 'que-faire-a-paris-')
        total_max (int): nombre total max de lignes à récupérer
        page_size (int): taille de page (max 100 imposé par l'API)
        timeout (int): timeout par requête en secondes

    Returns:
        pd.DataFrame ou None en cas d'échec
    """
    base_url = (
        f"https://opendata.paris.fr/api/explore/v2.1/"
        f"catalog/datasets/{dataset_id}/records"
    )
    page_size = min(page_size, 100)  # l'API limite à 100
    all_records = []

    try:
        for offset in range(0, total_max, page_size):
            limit = min(page_size, total_max - offset)
            params = {'limit': limit, 'offset': offset}

            response = requests.get(base_url, params=params, timeout=timeout)

            if response.status_code != 200:
                print(f"❌ Erreur HTTP {response.status_code} pour '{dataset_id}' "
                      f"(offset={offset})")
                print(f"   Détail : {response.text[:200]}")
                break

            data = response.json()
            records = data.get('results', [])

            if not records:
                # plus de résultats, on s'arrête
                break

            all_records.extend(records)

            # Si l'API renvoie moins que demandé, on a atteint la fin
            if len(records) < limit:
                break

        if not all_records:
            print(f"⚠️  Aucune donnée retournée pour '{dataset_id}'.")
            return None

        df = pd.DataFrame(all_records)
        print(f"✅ Dataset '{dataset_id}' chargé ({len(df)} lignes, "
              f"{df.shape[1]} colonnes).")
        return df

    except requests.exceptions.Timeout:
        print(f"⏱️  Timeout sur '{dataset_id}' après {timeout}s.")
        return None
    except requests.exceptions.RequestException as e:
        print(f"❌ Erreur de connexion : {e}")
        return None
    except ValueError as e:
        print(f"❌ Réponse JSON invalide : {e}")
        return None
```

```python
# =============================================================================
# 2. CHARGEMENT DES DEUX JEUX DE DONNÉES
# =============================================================================

print("🚀 Lancement du chargement des données depuis Open Data Paris (API v2.1)...\n")

df_sorties = charger_opendata_paris("que-faire-a-paris-", total_max=1000)
df_chantiers = charger_opendata_paris("chantiers-perturbants", total_max=1000)

if df_sorties is not None:
    print("\n📋 Colonnes 'sorties' :", df_sorties.columns.tolist()[:15], "...")
    display(df_sorties.head())

if df_chantiers is not None:
    print("\n📋 Colonnes 'chantiers' :", df_chantiers.columns.tolist())
    display(df_chantiers.head())
```

```python
# =============================================================================
# 3. EXPLORATION SÉCURISÉE DES SORTIES (ÉVÉNEMENTS)
# =============================================================================

if df_sorties is not None:
    print("\n--- 🎭 ANALYSE DES SORTIES À PARIS ---")

    # Colonnes qui nous intéressent (on garde celles qui existent réellement)
    cols_interet = ['title', 'category_label', 'address_zipcode', 'tags',
                    'date_start', 'date_end', 'address_name']
    cols_valides = [c for c in cols_interet if c in df_sorties.columns]

    print(f"Colonnes utilisées : {cols_valides}")
    if cols_valides:
        display(df_sorties[cols_valides].head())

    # Top 10 codes postaux pour les événements
    if 'address_zipcode' in df_sorties.columns:
        plt.figure(figsize=(10, 5))
        (df_sorties['address_zipcode']
            .dropna()
            .value_counts()
            .head(10)
            .plot(kind='bar', color='skyblue', edgecolor='black'))
        plt.title("Top 10 des arrondissements par nombre d'événements")
        plt.ylabel("Nombre d'activités")
        plt.xlabel("Code postal")
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
    else:
        print("⚠️  Colonne 'address_zipcode' absente — graphique sauté.")
```

```python
# =============================================================================
# 4. EXPLORATION SÉCURISÉE DES CHANTIERS (partie complétée)
# =============================================================================

if df_chantiers is not None:
    print("\n--- 🏗️ ANALYSE DES CHANTIERS PERTURBANTS ---")

    # Colonnes potentiellement présentes — l'API peut faire varier les noms,
    # on liste large et on filtre sur ce qui existe réellement.
    cols_chantier_interet = [
        'objet', 'libelle', 'nature', 'maitre_d_ouvrage',
        'niveau_perturbation', 'impact_circulation',
        'cp_arrondissement', 'arrondissement', 'code_postal',
        'date_debut', 'date_fin', 'voie', 'localisation_detaillee'
    ]
    cols_chantier_valides = [c for c in cols_chantier_interet
                             if c in df_chantiers.columns]

    print(f"Colonnes retenues : {cols_chantier_valides}")
    if cols_chantier_valides:
        display(df_chantiers[cols_chantier_valides].head())

    # --- Analyse de l'impact sur la circulation ---
    impact_col = next((c for c in ['impact_circulation', 'niveau_perturbation']
                       if c in df_chantiers.columns), None)

    if impact_col is not None:
        print(f"\n📊 Répartition de '{impact_col}' :")
        print(df_chantiers[impact_col].value_counts(dropna=False))

        plt.figure(figsize=(10, 5))
        order = df_chantiers[impact_col].value_counts().index
        sns.countplot(data=df_chantiers, y=impact_col,
                      order=order, palette='OrRd_r')
        plt.title(f"Répartition des chantiers selon : {impact_col}")
        plt.xlabel("Nombre de chantiers")
        plt.ylabel(impact_col)
        plt.tight_layout()
        plt.show()
    else:
        print("⚠️  Aucune colonne d'impact circulation trouvée.")

    # --- Répartition par arrondissement ---
    arr_col = next((c for c in ['cp_arrondissement', 'arrondissement', 'code_postal']
                    if c in df_chantiers.columns), None)

    if arr_col is not None:
        plt.figure(figsize=(12, 5))
        (df_chantiers[arr_col]
            .dropna()
            .astype(str)
            .value_counts()
            .head(15)
            .plot(kind='bar', color='salmon', edgecolor='black'))
        plt.title(f"Top 15 — chantiers par {arr_col}")
        plt.ylabel("Nombre de chantiers")
        plt.xlabel(arr_col)
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

    # --- Synthèse : durée des chantiers si dates disponibles ---
    if {'date_debut', 'date_fin'}.issubset(df_chantiers.columns):
        df_dates = df_chantiers[['date_debut', 'date_fin']].copy()
        df_dates['date_debut'] = pd.to_datetime(df_dates['date_debut'], errors='coerce')
        df_dates['date_fin'] = pd.to_datetime(df_dates['date_fin'], errors='coerce')
        df_dates['duree_jours'] = (df_dates['date_fin'] - df_dates['date_debut']).dt.days

        valid = df_dates['duree_jours'].dropna()
        if len(valid):
            print(f"\n📅 Durée des chantiers — médiane : {valid.median():.0f} jours, "
                  f"moyenne : {valid.mean():.0f} jours, max : {valid.max():.0f} jours")
```
